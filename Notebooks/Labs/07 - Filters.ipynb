## <font style="color:rgb(50,120,229)">Convolution</font>

Convolves an image with the kernel. [Link](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#ga27c049795ce870216ddfb366086b5a04)

The function applies an arbitrary linear filter to an image. When the aperture is partially outside the image, the function interpolates outlier pixel values according to the specified border mode.

The function does actually compute correlation, not the convolution:

````
kernel_size = 5
# Create a 5*5 kernel with all elements equal to 1
kernel = np.ones((kernel_size, kernel_size), dtype=np.float32) / kernel_size**2

# Print Kernel
print (kernel)

result = cv2.filter2D(img, -1, kernel, (-1, -1), delta=0, borderType=cv2.BORDER_DEFAULT)

# Display Output
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(img);plt.title("Original Image")
plt.subplot(122);plt.imshow(result);plt.title("Convolution Result : KernelSize = 5");
````

## <font style="color:rgb(50,120,229)">Gaussian noise</font>

````
# Read color image
DATA_PATH = '/content/drive/My Drive/Colab Notebooks/Computer Vision course/images/'
filename = DATA_PATH+"gaussian-noise.png"
img = plt.imread(filename)
plt.imshow(img)
````

#### <font style = "color:rgb(50,120,229)">Box Filter</font>
Blurs an image using the box filter. [Link](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#gad533230ebf2d42509547d514f7d3fbc3)

````
# Apply box filter - kernel size 3
dst1=cv2.boxFilter(img, -1,(3,3),(-1,-1))

# Apply box filter - kernel size 7
dst2=cv2.boxFilter(img,-1,(7,7),(-1,-1))

# Display Output
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(img);plt.title("Original Image")
plt.subplot(132);plt.imshow(dst1);plt.title("Box Blur Result 1 : KernelSize = 3")
plt.subplot(133);plt.imshow(dst2);plt.title("Box Blur Result 1 : KernelSize = 7");
````

#### <font style = "color:rgb(50,120,229)">Gaussian Filter</font>
Blurs an image using a Gaussian filter. [Link](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#gaabe8c836e97159a9193fb0b11ac52cf1)

The function convolves the source image with the specified Gaussian kernel.

````
# Apply gaussian blur
dst1=cv2.GaussianBlur(img,(5,5),0,0)
dst2=cv2.GaussianBlur(img,(25,25),50,50)

# Display Output
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(img);plt.title("Original Image")
plt.subplot(132);plt.imshow(dst1);plt.title("Gaussian Blur Result 1 : KernelSize = 5")
plt.subplot(133);plt.imshow(dst2);plt.title("Gaussian Blur Result 2 : KernelSize = 25");
````
